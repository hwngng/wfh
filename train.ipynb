```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer


def preprocessor(d):
	return d

def list_tokenizer(d):
	return d

label_set = ["dev", "dev", "dev", "dev", "test", "ba"]
f = open('train.json', 'r')
train_set = json.load(f)

app_set = [x['application'] for x in train_set]

tfidf = TfidfVectorizer(preprocessor=preprocessor, tokenizer=list_tokenizer)
app_set_tfidf = tfidf.fit_transform(app_set)
print(tfidf.vocabulary_)
```

```python
feature_names = tfidf.get_feature_names_out()
corpus_index = [i for i in range(len(app_set))]
import pandas as pd
pd.set_option('display.max_rows', 500)
app_set_tfidf
df = pd.DataFrame(app_set_tfidf.T.todense(), index=feature_names, columns=corpus_index)
print(df)
```
